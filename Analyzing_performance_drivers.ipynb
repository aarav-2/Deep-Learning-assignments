```python
#**************************************
#*** 4  Relations between variables ***
#**************************************

# Tell Python the structure of the panel data (ID and time variable), this is necessary to run panel regressions
# You cannot simply run an OLS regression with pooled data because the observations from one company in different years are not independent of each other
# (OLS regression treats all observations as having independent error terms)
df = df.set_index(['Company', 'Year'])
```

```python
# *****************************************
# *** Relation between customer satisfaction and sales ***
# *****************************************
from linearmodels.panel import RandomEffects



# The Random Effects model is being used here to estimate how customer satisfaction
# (MC5, i.e., % of customers who would recommend the company) influences sales (MF9).

# Define and fit the model using MC5 (customer satisfaction) as an independent variable
print(RandomEffects.from_formula('MF9 ~ 1 + MC5', data=df).fit().summary)

# Interpretation of the model:
# -> If you invest in initiatives that increase customer satisfaction (as measured by % of customers
# who would recommend the company, MC5), the model will estimate how that influences sales (MF9).
# -> The coefficient for MC5 will tell you the expected increase in sales (MF9) for each 1% increase
# in customer satisfaction (MC5).
# -> For example, if the coefficient for MC5 is 0.02, this means that for each 1% increase in the
# percentage of customers who would recommend the company, sales increase by 2%.
# -> Note that the analysis only shows a positive relationship between customer satisfaction and sales,
# not necessarily a direct causal relationship.
```

```python
# Lets do an OLS regression for comparison (using OLS is incorrect here, because the OLS regression assumes that all observations are independent from each other, which is not the case if we have multiple year-observations per firm)
import statsmodels.api as sm
import statsmodels.formula.api as smf
ols_model = smf.ols('MC5 ~ C5_MC5_S', data=df.reset_index()).fit()

# Print the summary of the model to see the results
print(ols_model.summary())
# -> In this particular case, the difference is not that big; however, in other cases, OLS and panel regressions can produce completely different results
```

```python
# ** Relation between customer ratings and sales **
# ** Is customer "Best in Class" ranking positively associated with sales?

print(RandomEffects.from_formula('MF9 ~ 1 + MC5', data=df).fit().summary)

# -> yes! If one percent more customers rank you as best in class, this increases revenue by 5.9 Million.
```

```python
#*************************
#*** 5  Non-linear effects ***
#*************************

# ** Non-linear effects
df['MC5_sq'] = df['MC5'] ** 2

# Panel regression with non-linear terms
print(RandomEffects.from_formula('MF9 ~ 1 + MC5 + MC5_sq', data=df).fit().summary)
```

```python
# *************************
# *** 6  Lagged effects ***
# *************************

# lagged effects
df= df.sort_values(by=['Company', 'Year'])

df['MC5_Lag1'] = df.groupby('Company')['MC5'].shift(1)
df['MC5_Lag2'] = df.groupby('Company')['MC5'].shift(2)
df['MC5_Lag3'] = df.groupby('Company')['MC5'].shift(3)
df['MC5_Lag4'] = df.groupby('Company')['MC5'].shift(4)
df['MC5_Lag5'] = df.groupby('Company')['MC5'].shift(5)


# Define and fit the models
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_Lag1', data=df).fit().summary)
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_Lag2', data=df).fit().summary)
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_Lag3', data=df).fit().summary)
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_Lag4', data=df).fit().summary)
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_Lag5', data=df).fit().summary)
```

```python
# *****************************************
# *** 7  Moderating/interaction effects ***
# *****************************************

# *****************************************
# *** Moderating effect (cross-sectional variation) ***
# *****************************************

# Create a 2-period lagged variable for MC5 (Customer Satisfaction)
df['MC5_lag_2'] = df.groupby('Company')['MC5'].shift(2)

# Create a binary variable for Product Innovation strategy
df['Product_innovation'] = (df['strategy'] == "Product innovation").astype(int)

# Interaction term between lagged customer satisfaction and the Product Innovation strategy
df['interaction'] = df['MC5_lag_2'] * df['Product_innovation']

# Filter the data for the 'Product innovation' strategy
df_product_innovation = df[df['strategy'] == "Product innovation"].dropna(subset=['MC5_lag_2'])

# Define and fit the Random Effects model
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_lag_2', data=df_product_innovation).fit().summary)

# Interpretation of the Model:
# -> This model estimates how customer satisfaction (lagged by 2 periods) affects sales (MF9),
# specifically for companies following the Product Innovation strategy.
# -> The coefficient for MC5_lag_2 will indicate how a change in customer satisfaction two periods ago
# influences sales in the current period.
# -> If the interaction term is used in future models, it will help assess if the effect of customer satisfaction
# differs depending on whether or not the company follows the Product Innovation strategy.
```

```python

# Create a 2-period lagged variable for MC5 (Customer Satisfaction)
df['MC5_lag_2'] = df.groupby('Company')['MC5'].shift(2)

# Create a binary variable for Customer Integration strategy
df['Customer_integration'] = (df['strategy'] == "Customer integration").astype(int)

# Interaction term between lagged customer satisfaction and the Customer Integration strategy
df['interaction'] = df['MC5_lag_2'] * df['Customer_integration']

# Filter the data for the 'Customer integration' strategy
df_Customer_integration = df[df['strategy'] == "Customer integration"].dropna(subset=['MC5_lag_2'])

# Define and fit the Random Effects model using both lagged and current customer satisfaction (MC5)
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_lag_2 + MC5', data=df_Customer_integration).fit().summary)

# Interpretation of the Model:
# -> This model estimates how both lagged customer satisfaction (MC5_lag_2, two periods ago) and current customer satisfaction (MC5)
# affect sales (MF9), specifically for companies following the Customer Integration strategy.
# -> The coefficients for MC5_lag_2 and MC5 will show how customer satisfaction from two periods ago and the current period,
# respectively, influence sales.
# -> The interaction term can be used to assess whether the effect of lagged customer satisfaction differs under the Customer Integration strategy.
```

```python
# *****************************************
# *** Moderating effect (cross-sectional variation) ***
# *****************************************

# Create a 2-period lagged variable for MC5 (Customer Satisfaction)
df['MC5_lag_2'] = df.groupby('Company')['MC5'].shift(2)

# Create a binary variable for Low Lifetime Cost strategy
df['Low_lifetime_cost'] = (df['strategy'] == "Low lifetime cost").astype(int)

# Interaction term between lagged customer satisfaction and the Low Lifetime Cost strategy
df['interaction'] = df['MC5_lag_2'] * df['Low_lifetime_cost']

# Filter the data for the 'Low lifetime cost' strategy
df_Low_lifetime_cost = df[df['strategy'] == "Low lifetime cost"].dropna(subset=['MC5_lag_2'])

# Define and fit the Random Effects model using both lagged and current customer satisfaction (MC5)
print(RandomEffects.from_formula('MF9 ~ 1 + MC5_lag_2 + MC5', data=df_Low_lifetime_cost).fit().summary)

# Interpretation of the Model:
# -> This model estimates how both lagged customer satisfaction (MC5_lag_2, two periods ago) and current customer satisfaction (MC5)
# affect sales (MF9), specifically for companies following the Low Lifetime Cost strategy.
# -> The coefficients for MC5_lag_2 and MC5 will show how customer satisfaction from two periods ago and the current period,
# respectively, influence sales.
# -> The interaction term can be used to assess whether the effect of lagged customer satisfaction differs under the Low Lifetime Cost strategy.
```
